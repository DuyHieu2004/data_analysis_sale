```python
import pandas as pd

# Đọc dữ liệu.
df = pd.read_csv('retail_sales_dataset_extended.csv')
df.head(5)
```

```python
# Kiểm tra thông tin.
print(df.info())
print(df.isnull().sum())
```

```python
# Kiểm tra giá trị bất thường.
print('Kiểm tra Total Amount:')
df['Check_Total_Amount'] = df['Quantity'] * df['Price per Unit']
print('\n')
print(df[df['Check_Total_Amount'] != df['Total Amount']])
```

```python
# Chuẩn hóa dữ liệu.
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

df.head(5)
```

```python
# Phân nhóm tuổi.
bins = [18, 25, 35, 50, 100]
labels = ['18-25', '26-35', '36-50', '>50']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

df.head(5)
```

```python
# Lưu dữ liệu đã được làm sạch.
df.to_csv('retail_sales_dataset_cleaned.csv', index=False)
print("Dữ liệu đã được làm sạch và lưu vào 'retail_sales_dataset_cleaned.csv'")
```
